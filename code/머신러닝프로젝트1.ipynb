#### 이 자료는 한해의 매출액을 나타낸 자료이기 때문에 한 해의 후반기에 개봉되었을 경우 매출액이 낮게 측정 될 수가 있다
#### 따라서 같은 영화일 경우 매출액을 더해서 하나의 데이터로 만들어 줘야함

```python
duple_list = list(df.duplicated(subset=['영화명', '러닝타임'], keep='first'))
```

```python
duple_list2 = df[duple_list]['영화명']
```

```python
df2 = df.copy()
df2.head()
```

```python
sum(df2.loc[df2['영화명']=='가면', '매출액'])
```

```python
df2[df2['영화명']=='가면']
```

```python
df2.drop(columns = '매출액 점유율')
```

```python
# 같은 영화의 매출액을 합으로 수정
for n in duple_list2:
    df2.loc[df2['영화명']== n, '매출액'] = sum(df2.loc[df2['영화명']==n, '매출액'])
    df2.loc[df2['영화명']== n, '관객수'] = sum(df2.loc[df2['영화명']==n, '관객수'])
    df2.loc[df2['영화명']== n, '스크린수'] = sum(df2.loc[df2['영화명']==n, '스크린수'])
    df2.loc[df2['영화명']== n, '상영횟수'] = sum(df2.loc[df2['영화명']==n, '상영횟수'])
```
