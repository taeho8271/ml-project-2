```python
title_len = []
count = 0
for name in movie_df['영화명']:
    str_name = str(name)
    sum = 0
    count += 1 
    
    
    for i in str_name.replace(' ', ''):
        sum += len(i)
    
    title_len.append(sum)
    
```

```python
len(title_len),count,title_len
```

```python
movie_df['영화제목_길이'] = title_len
```
