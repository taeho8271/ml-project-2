```python
import pandas as pd


new_encoding = pd.read_excel('./새로운라벨인코딩(3).xlsx', index_col=0)
new_encoding
```

```python
# drop 할 컬럼 확인 

new_encoding.info()
```

```python
new_encoding = new_encoding.astype({'영화명':'str'})
```

```python
# 영화명 인코딩 하기

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le.fit(new_encoding['영화명'])

new_encoding['le_영화명'] = le.transform(new_encoding['영화명'])
```

```python
# boxplot 확인하기
# 상영횟수만 너무 범위가 큼 >> 일단 넣고 성능 테스트 후, 나중에 와서 상영횟수만 스케일링 해서 넣어 본 후 성능 테스트 해보기

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
# plt.xlabel( labelpad=30)
plt.ylabel('Fetures', labelpad=30, loc='top')
plt.rc('xtick', labelsize= 10)
# plt.xticks(rotation=90)

sns.boxplot(data=new_encoding.drop(['take'],axis=1), orient='h')
```

```python
# 일단 성능 검사 먼저

# 데이터 나누기

from sklearn.model_selection import train_test_split

X = new_encoding.drop(['take'],axis=1)
y = new_encoding['take']

X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=13)
```

```python
X_train.shape , X_test.shape
```

```python
# 머신러닝 만들기

# linear regression 모델 사용하기
# label을 붙여 학습시키고, output은 연속적인 값을 예상해야 하기 때문에

from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train,y_train) # 학습
```

```python
# 모델 평가하기 # rmse


import numpy as np
from sklearn.metrics import mean_squared_error

pred_tr = reg.predict(X_train)
pred_test = reg.predict(X_test)

rmse_tr = np.sqrt(mean_squared_error(y_train, pred_tr))
rmse_test = np.sqrt(mean_squared_error(y_test, pred_test))


print('RMSE of Train data : ', rmse_tr)
print('RMSE of Test data : ', rmse_test)
```

```python
# 성능 확인

plt.scatter(y_test, pred_test, color='lightpink') # 실제값을 넣었을때 예상한 예상값
plt.xlabel('actual movie take')
plt.ylabel('Predicted take')
plt.title("Real vs Predicted")

plt.plot([min(y_test),max(y_test)],[min(y_test),max(y_test)], 'b', lw=3, ls='dashed') # 실제 값
plt.show()
```

```python
# 스크린 상영 횟수 스케일링 해보기

from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

new_encoding['ss_number of screenings'] = ss.fit_transform(new_encoding['number of screenings'].values.reshape(-1,1))
```

```python
new_encoding = new_encoding.drop(['number of screenings'], axis=1)
```

```python
# boxplot 재확인하기


import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
# plt.xlabel( labelpad=30)
plt.ylabel('Fetures', labelpad=30, loc='top')
plt.rc('xtick', labelsize= 10)
# plt.xticks(rotation=90)

sns.boxplot(data=new_encoding.drop(['take'],axis=1) ,orient='h')
```
